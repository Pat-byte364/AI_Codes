```python
class DSU:
    def __init__(self, n):
        self.parent = list(range(n))

    def find(self, x):
        if self.parent[x] != x:
            self.parent[x] = self.find(self.parent[x])
        return self.parent[x]

    def union(self, x, y):
        px = self.find(x)
        py = self.find(y)

        if px == py:
            return False   # cycle

        self.parent[px] = py
        return True


def kruskal(n, edges):
    edges.sort(key=lambda x: x[2])  # sort by weight

    dsu = DSU(n)
    mst = []
    total = 0

    for u, v, w in edges:
        if dsu.union(u, v):
            mst.append((u, v, w))
            total += w

    return mst, total


# Driver
edges = [
    (0,1,10),
    (0,2,6),
    (0,3,5),
    (1,3,15),
    (2,3,4)
]

mst, cost = kruskal(4, edges)

print("MST:", mst)
print("Total Cost:", cost)
```
