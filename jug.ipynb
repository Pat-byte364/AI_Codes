```python
def water_jug_dfs(capA, capB, target):
    visited = set()
    path = []

    def dfs(A, B):
        # Goal: target in any jug
        if A == target or B == target:
            path.append((A, B))
            return True

        if (A, B) in visited:
            return False

        visited.add((A, B))
        path.append((A, B))

        # All possible moves
        moves = [
            (capA, B),                     # Fill Jug A
            (A, capB),                     # Fill Jug B
            (0, B),                        # Empty Jug A
            (A, 0),                        # Empty Jug B
            # Pour A -> B
            (A - min(A, capB - B), B + min(A, capB - B)),
            # Pour B -> A
            (A + min(B, capA - A), B - min(B, capA - A))
        ]

        for next_A, next_B in moves:
            if dfs(next_A, next_B):
                return True

        path.pop()   # Backtrack
        return False

    dfs(0, 0)
    return path


# Driver code
solution = water_jug_dfs(4, 3, 2)

print("Solution path:")
for step, (A, B) in enumerate(solution):
    print(f"Step {step}: Jug A = {A}, Jug B = {B}")
```
