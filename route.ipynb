```python
class Route:
    def __init__(self, start, end, distance, time, cost):
        self.start = start
        self.end = end
        self.distance = distance
        self.time = time
        self.cost = cost
        self.left = None
        self.right = None

    def get_value(self, criterion):
        if criterion == "distance":
            return self.distance
        elif criterion == "time":
            return self.time
        else:
            return self.cost


class BST:
    def __init__(self, criterion):
        self.root = None
        self.criterion = criterion

    def insert(self, node, route):
        if node is None:
            return route

        if route.get_value(self.criterion) < node.get_value(self.criterion):
            node.left = self.insert(node.left, route)
        else:
            node.right = self.insert(node.right, route)

        return node

    def find_best(self):
        current = self.root
        while current.left:
            current = current.left
        return current


# 🔥 Driver Code
routes = [
    Route("A", "B", 350, 240, 45),
    Route("A", "C", 260, 160, 25),
    Route("B", "D", 450, 280, 35),
    Route("C", "D", 610, 380, 80)
]

bst = BST("distance")   # change to "time" or "cost"

for r in routes:
    bst.root = bst.insert(bst.root, r)

best = bst.find_best()

print("Best Route:")
print(best.start, "->", best.end, "| Distance:", best.distance)
```
