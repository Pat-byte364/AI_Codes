```python
import heapq

def heuristic(state, goal):
    dist = 0
    for i in range(len(state)):
        if state[i] != 0:
            goal_index = goal.index(state[i])
            dist += abs(i//3 - goal_index//3) + abs(i%3 - goal_index%3)
    return dist

def astar(start, goal):
    pq = []
    heapq.heappush(pq, (0, 0, start, []))  # (f, g, state, path)

    visited = set()

    while pq:
        f, g, state, path = heapq.heappop(pq)

        if state == goal:
            return path + [state]

        if state in visited:
            continue
        visited.add(state)

        zero = state.index(0)
        row, col = zero//3, zero%3

        moves = [(-1,0),(1,0),(0,-1),(0,1)]

        for dr, dc in moves:
            r, c = row+dr, col+dc
            if 0<=r<3 and 0<=c<3:
                new = list(state)
                idx = r*3 + c
                new[zero], new[idx] = new[idx], new[zero]
                new = tuple(new)

                new_g = g + 1
                h = heuristic(new, goal)

                heapq.heappush(pq, (new_g + h, new_g, new, path + [state]))

    return None

start = (1,2,3,
         4,0,6,
         7,5,8)

goal = (1,2,3,
        4,5,6,
        7,8,0)

result = astar(start, goal)

print("Solution path:")
for step in result:
    print(step)
```
