```python
from collections import deque

def bfs(start, goal):
    queue = deque([(start, [])])
    visited = set([start])

    while queue:
        state, path = queue.popleft()

        if state == goal:
            return path + [state]

        zero = state.index(0)
        row, col = zero // 3, zero % 3

        moves = [(-1,0),(1,0),(0,-1),(0,1)]

        for dr, dc in moves:
            r, c = row + dr, col + dc
            if 0 <= r < 3 and 0 <= c < 3:
                new = list(state)
                idx = r*3 + c
                new[zero], new[idx] = new[idx], new[zero]
                new_state = tuple(new)

                if new_state not in visited:
                    visited.add(new_state)
                    queue.append((new_state, path + [state]))

    return None


# Driver
start = (1,2,3,
         4,0,6,
         7,5,8)

goal = (1,2,3,
        4,5,6,
        7,8,0)

res = bfs(start, goal)

for step in res:
    print(step)
```
