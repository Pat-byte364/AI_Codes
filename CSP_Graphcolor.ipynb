```python
def is_safe(node, color, colors, graph):
    for neighbor in graph[node]:
        if colors[neighbor] == color:
            return False
    return True


def graph_coloring(graph, m, colors, node):
    # If all nodes are colored
    if node == len(graph):
        return True

    # Try all colors
    for c in range(1, m + 1):
        if is_safe(node, c, colors, graph):
            colors[node] = c

            if graph_coloring(graph, m, colors, node + 1):
                return True

            # Backtrack
            colors[node] = 0

    return False


# GRAPH (your diagram)
graph = {
    0: [1, 2],
    1: [0, 2, 3],
    2: [0, 1, 3],
    3: [1, 2]
}

n = len(graph)
colors = [0] * n
m = 3  # number of colors

if graph_coloring(graph, m, colors, 0):
    print("Solution:", colors)
else:
    print("No solution")
```
