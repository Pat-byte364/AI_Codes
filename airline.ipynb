```python
import datetime

class Flight:
    def __init__(self, id, origin, dest, dep, duration):
        self.id = id
        self.origin = origin
        self.dest = dest
        self.dep = dep
        self.arr = dep + datetime.timedelta(minutes=duration)

class Aircraft:
    def __init__(self, name, location):
        self.name = name
        self.location = location
        self.available = datetime.datetime.min

    def can_assign(self, flight):
        return self.location == flight.origin and self.available <= flight.dep

    def assign(self, flight):
        self.location = flight.dest
        self.available = flight.arr

def schedule(flights, aircrafts):

    for f in flights:
        assigned = False

        for a in aircrafts:
            if a.can_assign(f):
                a.assign(f)
                print(f"{f.id} -> {a.name}")
                assigned = True
                break

        if not assigned:
            print(f"{f.id} -> Not Assigned")


# Driver
t = datetime.datetime(2024,1,1,6,0)

flights = [
    Flight("F1","A","B", t+datetime.timedelta(hours=1),120),
    Flight("F2","B","C", t+datetime.timedelta(hours=4),90),
]

aircrafts = [
    Aircraft("Plane1","A")
]

schedule(flights, aircrafts)
```
